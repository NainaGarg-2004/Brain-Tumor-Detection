Smoothening and sharpening

```python
def preprocess_image(img):
    # Apply Gaussian Blurring for smoothing
    blurred_img = cv2.GaussianBlur(img, (5, 5), 0)  # You can adjust the kernel size (e.g., (5, 5)) based on your preference
    
    # Apply Unsharp Masking for sharpening
    sharpened_img = cv2.addWeighted(img, 3, blurred_img, -0.9, 0)
    
    return sharpened_img
```

```python
import cv2
X = []
Y = []
for cls in classes:
    pth = 'C:/Users/91935/Documents/Brain_Tumor/Training/'+cls
    for j in os.listdir(pth):
        img = cv2.imread(pth+'/'+j, 0)
        img = cv2.resize(img, (200,200))
        X.append(img)
        Y.append(classes[cls])
```

```python
img_path = 'C:/Users/91935/Documents/Brain_Tumor/Testing/no_tumor/image(37).jpg'  # Change this to the path of your image
img = cv2.imread(img_path, 0)  # Read the image in grayscale
if img is None:
    print("Error: Unable to read the image.")
else:
    preprocessed_img = preprocess_image(img)
    plt.figure(figsize=(8, 8))
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.imshow(preprocessed_img, cmap='BrBG_r')
    plt.title('Preprocessed Image')
    plt.axis('off')
    plt.show()
```

### Prepare data

```python
X_updated = X.reshape(len(X), -1)
X_updated.shape
```

### Split Data

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_updated, Y, random_state=10,
                                               test_size=.20)
```

```python
xtrain.shape, xtest.shape
```

### Feature Scaling

```python
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())
xtrain = xtrain/255
xtest = xtest/255
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())
```

### Feature Selection: PCA

```python
from sklearn.decomposition import PCA
```

```python
print(xtrain.shape, xtest.shape)

pca = PCA(.98)
# pca_train = pca.fit_transform(xtrain)
# pca_test = pca.transform(xtest)
pca_train = xtrain
pca_test = xtest
```

```python
# print(pca_train.shape, pca_test.shape)
# print(pca.n_components_)
# print(pca.n_features_)
```

### Train Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
import warnings
warnings.filterwarnings('ignore')

lg = LogisticRegression(C=0.1)
lg.fit(pca_train, ytrain)
```

```python
sv = SVC()
sv.fit(pca_train, ytrain)
```

### Evaluation

```python
print("Training Score:", lg.score(pca_train, ytrain))
print("Testing Score:", lg.score(pca_test, ytest))
```

```python
print("Training Score:", sv.score(pca_train, ytrain))
print("Testing Score:", sv.score(pca_test, ytest))
```
